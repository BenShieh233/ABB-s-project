#### 1. Customers

```python
def customers(company_id):
    global username, password
    url = 'https://www.capitaliq.com/CIQDotNet/BusinessRel/Customers.aspx?CompanyId={}'
    bot = webdriver.Chrome(options=options)
    wait = WebDriverWait(bot, 20)
    bot.get(url.format(company_id))
    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "#onetrust-accept-btn-handler"))).click()

    time.sleep(1)

    bot.find_element(By.ID, 'input28').send_keys(username)
    bot.find_element(By.ID, 'input28').send_keys(Keys.RETURN)

    time.sleep(1)

    # Assuming 'bot' is your WebDriver instance
    password_field = bot.find_element(By.ID, 'input60')
    password_field.send_keys(password)
    password_field.send_keys(Keys.RETURN)

    time.sleep(1)

    table = bot.find_element(By.ID, "myCustomersGrid_gridSection_myDataGrid")
    data = []
    for row in table.find_elements(By.TAG_NAME, "tr"):
        row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
        if row_data:
            data.append(row_data)
    headers = data[0]
    non_empty_headers = [header for header in headers if header]
    cleaned_rows = [[cell for cell in row if cell] for row in data[1:]]
    cleaned_rows = [row for row in cleaned_rows if any(cell for cell in row if cell)]
    df = pd.DataFrame(cleaned_rows[1:], columns=non_empty_headers)
    df.insert(0, 'company_id', company_id)

    df = df[:-1]
 
    bot.quit()
    return df
```

#### 7. Geographical Presence

```python
def geo_segment(company_id):
    global username, password
    url = 'https://www.capitaliq.com/CIQDotNet/Financial/Segments.aspx?CompanyId={}'
    bot = webdriver.Chrome(options=options)
    wait = WebDriverWait(bot, 20)
    bot.get(url.format(company_id))
    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "#onetrust-accept-btn-handler"))).click()

    time.sleep(1)

    bot.find_element(By.ID, 'input28').send_keys(username)
    bot.find_element(By.ID, 'input28').send_keys(Keys.RETURN)

    time.sleep(1)

    # Assuming 'bot' is your WebDriver instance
    password_field = bot.find_element(By.ID, 'input60')
    password_field.send_keys(password)
    password_field.send_keys(Keys.RETURN)

    time.sleep(1)

    bs_table = bot.find_element(By.ID, "__gv_dispSection_BS_ctl00_0_gv_0__div")
    data = []
    for row in bs_table.find_elements(By.TAG_NAME, "tr"):
        row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
        if row_data:
            data.append(row_data)
    headers = data[0]
    non_empty_headers = [header for header in headers if header]
    cleaned_rows = [[cell for cell in row if cell] for row in data[1:]]
    cleaned_rows = [row for row in cleaned_rows if any(cell for cell in row if cell)]
    df_bs = pd.DataFrame(cleaned_rows[1:], columns=non_empty_headers)

    df_bs.insert(0, 'company_id', company_id)


    gs_table = bot.find_element(By.ID, "__gv_dispSection_GS_ctl00_0_gv_0__div")

    data = []
    for row in gs_table.find_elements(By.TAG_NAME, "tr"):
        row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
        if row_data:
            data.append(row_data)
    headers = data[0]
    non_empty_headers = [header for header in headers if header]
    cleaned_rows = [[cell for cell in row if cell] for row in data[1:]]
    cleaned_rows = [row for row in cleaned_rows if any(cell for cell in row if cell)]
    df_gs = pd.DataFrame(cleaned_rows[1:], columns=non_empty_headers)

    df_gs.insert(0, 'company_id', company_id)

    bot.quit()

    return df_bs, df_gs
```
