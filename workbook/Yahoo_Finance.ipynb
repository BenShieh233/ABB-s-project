```python
from numpy import True_
from pickle import TRUE
from re import I
import yfinance as yf
import pandas as pd

# List of tickers for the companies you want to get financials for
tickers = ['ADM','BUD','BGS','BYND','BF-B','BG','CPB','CL','CAG','CTVA', 'KDSKF','GIS','INGR','JBSAY','KHC','STZ','MDLZ','NSRGY','OLGPF','PEP', 'SAP.TO','TATE.L','TSN','UL','SJM']

df_list = []
for ticker_symbol in tickers:
    ticker = yf.Ticker(ticker_symbol)
    df = ticker.financials  # Make sure to call the correct attribute for financials

    # Attempt to fetch standard financial statement rows, some may not be available for all companies
    index_vars = ["Total Revenue", "Gross Profit", "Operating Income", "Operating Revenue", "Restructuring And Mergern Acquisition","Operating Expense"]
    # Adjust to the correct field names you expect from your data source

    # Get the current columns as years
    col_years = [col.year for col in df.columns]

    # Renaming the columns to years might not be necessary if you use them directly
    df.rename({col: col_year for col, col_year in zip(df.columns, col_years)}, axis=1, inplace=True)

    # Check if the desired rows exist in the DataFrame and filter the columns by the desired years
    existing_index_vars = [var for var in index_vars if var in df.index]
    years = [2023,2022, 2021, 2020]  # Use the actual years you want to filter
    df = df.loc[existing_index_vars, df.columns.isin(years)]

    # Add ticker symbol to the DataFrame
    df['Ticker'] = ticker_symbol

    df_list.append(df)

# Concatenate all DataFrames into one
financial_df = pd.concat(df_list, axis=0)
financial_df.head(1000)
financial_df.reset_index(names=['financial_variables'],inplace=True)

```

```python
df_list_1 = []
for ticker_symbol in tickers:
    ticker = yf.Ticker(ticker_symbol)
    df1 = ticker.balance_sheet  # Make sure to call the correct attribute for financials

    # Attempt to fetch standard financial statement rows, some may not be available for all companies
    index_vars_1 = ["Total Debt", "Current Assets", "Current Liabilities"]
    # Adjust to the correct field names you expect from your data source

    # Get the current columns as years
    col_years_1 = [col.year for col in df1.columns]

    # Renaming the columns to years might not be necessary if you use them directly
    df1.rename({col: col_year_1 for col, col_year_1 in zip(df1.columns, col_years_1)}, axis=1, inplace=True)

    # Check if the desired rows exist in the DataFrame and filter the columns by the desired years
    existing_index_vars_1 = [var for var in index_vars_1 if var in df1.index]
    years_1 = [2023,2022, 2021, 2020]  # Use the actual years you want to filter
    df1 = df1.loc[existing_index_vars_1, df1.columns.isin(years_1)]
    # Add ticker symbol to the DataFrame
    df1['Ticker'] = ticker_symbol

    df_list_1.append(df1)

# Concatenate all DataFrames into one
financial_df_1 = pd.concat(df_list_1, axis=0)
financial_df_1.head(1000)
financial_df_1.reset_index(names=['financial_variables'],inplace=True)

#financial_df.T

pd.melt(financial_df_1, id_vars=["Ticker","financial_variables"], value_vars=[2023, 2022,2021,2020],var_name="year")
```

```python
# Create a dictionary to store the extracted information
info_dict = {'Ticker': [], 'FinancialCurrency': [], 'RevenueGrowth': [],
             'GrossMargins': [], 'CurrentRatio': [],
             'RecommendationMean': [], 'RecommendationKey': [],
             'MarketCap': []}

for ticker_symbol in tickers:
    ticker = yf.Ticker(ticker_symbol)

    # Access specific variables from the ticker.info dictionary
    financial_currency = ticker.info.get('financialCurrency', None)
    revenue_growth = ticker.info.get('revenueGrowth', None)
    gross_margins = ticker.info.get('grossMargins', None)
    current_ratio = ticker.info.get('currentRatio', None)
    recommendation_mean = ticker.info.get('recommendationMean', None)
    recommendation_key = ticker.info.get('recommendationKey', None)
    market_cap = ticker.info.get('marketCap', None)

    # Append the values to the dictionary
    info_dict['Ticker'].append(ticker_symbol)
    info_dict['FinancialCurrency'].append(financial_currency)
    info_dict['RevenueGrowth'].append(revenue_growth)
    info_dict['GrossMargins'].append(gross_margins)
    info_dict['CurrentRatio'].append(current_ratio)
    info_dict['RecommendationMean'].append(recommendation_mean)
    info_dict['RecommendationKey'].append(recommendation_key)
    info_dict['MarketCap'].append(market_cap)

# Create a DataFrame from the dictionary
info_df = pd.DataFrame(info_dict)

# Display the resulting DataFrame
print(info_df)
```

```python
#!pip install forex-python
from forex_python.converter import CurrencyRates
# Use forex-python to get conversion rates
c = CurrencyRates()

financial_df = pd.merge(financial_df, info_df[['Ticker', 'FinancialCurrency']], how='left', on='Ticker')
financial_df_1 = pd.merge(financial_df_1, info_df[['Ticker', 'FinancialCurrency']], how='left', on='Ticker')

# Add 'Currency_to_USD' column
financial_df['Currency_to_USD'] = financial_df['FinancialCurrency'].apply(lambda x: c.get_rate(x, 'USD'))
financial_df_1['Currency_to_USD'] = financial_df_1['FinancialCurrency'].apply(lambda x: c.get_rate(x, 'USD'))

# Convert financial values to USD for each year
for year in [2022, 2021, 2020, 2023]:
    financial_df[f'Value_{year}_USD'] = financial_df[year] * financial_df['Currency_to_USD']
    financial_df_1[f'Value_{year}_USD'] = financial_df_1[year] * financial_df_1['Currency_to_USD']

# Display the resulting DataFrames
print(financial_df.head(1000))
print(financial_df_1.head(1000))
```

```python
# Calculate the sum of values for each financial variable
sum_values = filtered_df.groupby('financial_variables')['value'].sum()
sum_values
```

```python
# Calculate the score for each ticker and financial variable
filtered_df['score'] = filtered_df.apply(lambda row: row['value'] / sum_values[row['financial_variables']], axis=1)
print(filtered_df)
```
